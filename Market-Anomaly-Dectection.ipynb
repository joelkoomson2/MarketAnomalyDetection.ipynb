```python
# anomaly_detection_strategy.py

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.tree import export_graphviz

import matplotlib.pyplot as plt
import joblib

# -------------------------------
# STEP 1: Smart Time Split
# -------------------------------

def create_smart_time_split(df: pd.DataFrame):
    """
    Split time series data into train, validation, test sets for robust anomaly detection.
    """
    df['Data'] = pd.to_datetime(df['Data'])
    df = df.sort_values('Data').reset_index(drop=True)
    df['Year'] = df['Data'].dt.year

    train_data = df[df['Year'].between(2000, 2003)].copy()
    validation_data = df[df['Year'] == 2008].copy()
    test_data = df[df['Year'] == 2009].copy()

    print("\n=== Periods ===")
    print(f"Train: {train_data['Data'].min().date()} to {train_data['Data'].max().date()}")
    print(f"Val:   {validation_data['Data'].min().date()} to {validation_data['Data'].max().date()}")
    print(f"Test:  {test_data['Data'].min().date()} to {test_data['Data'].max().date()}")

    print("\n=== Class Balance ===")
    for split, data in zip(['Train', 'Val', 'Test'], [train_data, validation_data, test_data]):
        dist = data['Y'].value_counts(normalize=True).round(3).to_dict()
        print(f"{split}: {dist}")

    return train_data, validation_data, test_data


# -------------------------------
# STEP 2: Build and Tune Model Pipeline
# -------------------------------

def build_and_tune_pipeline(X_train, y_train, model_name="rf"):
    """
    Create pipeline, tune hyperparameters using grid search, and return best estimator.
    """
    if model_name == "rf":
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('rf', RandomForestClassifier(class_weight='balanced', random_state=42))
        ])
        param_grid = {
            'rf__n_estimators': [100, 300],
            'rf__max_depth': [5, 10, None],
            'rf__min_samples_leaf': [1, 5]
        }
    elif model_name == "logreg":
        pipe = Pipeline([
            ('scaler', StandardScaler()),
            ('logreg', LogisticRegression(
                class_weight='balanced',
                solver='liblinear',
                max_iter=1000,
                random_state=42
            ))
        ])
        param_grid = {
            'logreg__C': [0.01, 0.1, 1.0, 10]
        }
    else:
        raise ValueError("Unsupported model")

    grid = GridSearchCV(pipe, param_grid, scoring='roc_auc', cv=3, n_jobs=-1)
    grid.fit(X_train, y_train)

    print(f"\nBest Params ({model_name}): {grid.best_params_}")
    return grid.best_estimator_


# -------------------------------
# STEP 3: Evaluate Model
# -------------------------------

def evaluate_model(model, X_test, y_test):
    """
    Print classification metrics and AUC.
    """
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print("\n=== Classification Report ===")
    print(classification_report(y_test, y_pred))
    print(f"AUC: {roc_auc_score(y_test, y_prob):.4f}")

    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix:")
    print(cm)


# -------------------------------
# STEP 4: Investment Strategy Sketch
# -------------------------------

def backtest_strategy(predictions, returns, threshold=0.5):
    """
    Basic strategy: if model predicts crash probability > threshold, exit market.
    Returns is a pandas Series of daily returns.
    """
    positions = np.where(predictions > threshold, 0, 1)  # 0: out, 1: in
    strategy_returns = positions * returns

    cum_return = (strategy_returns + 1).cumprod()
    cum_bench = (returns + 1).cumprod()

    plt.figure(figsize=(12, 6))
    plt.plot(cum_return, label='Strategy')
    plt.plot(cum_bench, label='Benchmark (Buy & Hold)')
    plt.title("Backtest: Strategy vs Buy & Hold")
    plt.legend()
    plt.show()

    print(f"Strategy Final Return: {cum_return.iloc[-1]:.2f}")
    print(f"Benchmark Final Return: {cum_bench.iloc[-1]:.2f}")


# -------------------------------
# MAIN EXECUTION
# -------------------------------

if __name__ == "__main__":
    # Load data
    df = pd.read_csv("FinancialMarketData.csv")
    df.columns = df.columns.str.strip()

    train, val, test = create_smart_time_split(df)

    selected_features = ['VIX', 'DXY']  # ✅ Adjust this for your dataset!
    X_train = pd.concat([train, val])[selected_features]
    y_train = pd.concat([train, val])['Y']
    X_test = test[selected_features]
    y_test = test['Y']

    # Train models
    rf_model = build_and_tune_pipeline(X_train, y_train, model_name="rf")
    logreg_model = build_and_tune_pipeline(X_train, y_train, model_name="logreg")

    # Evaluate
    print("\nRandom Forest:")
    evaluate_model(rf_model, X_test, y_test)
    print("\nLogistic Regression:")
    evaluate_model(logreg_model, X_test, y_test)

    # Save best model
    joblib.dump(rf_model, "rf_model.pkl")
    joblib.dump(logreg_model, "logreg_model.pkl")

    # Example feature importances
    if hasattr(rf_model.named_steps['rf'], 'feature_importances_'):
        importances = rf_model.named_steps['rf'].feature_importances_
        plt.barh(selected_features, importances)
        plt.title("Random Forest Feature Importances")
        plt.show()

    if hasattr(logreg_model.named_steps['logreg'], 'coef_'):
        coef = logreg_model.named_steps['logreg'].coef_[0]
        plt.barh(selected_features, coef)
        plt.title("Logistic Regression Coefficients")
        plt.show()

    # Simulate strategy
    # Example: suppose daily returns column exists
    test['predicted_prob'] = rf_model.predict_proba(X_test)[:, 1]
    backtest_strategy(test['predicted_prob'], test['Returns'], threshold=0.5)
```
